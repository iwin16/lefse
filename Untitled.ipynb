```python
import pandas as pd
import os
```

```python
#reading file and making dataframes of each category
name_list = ["fw_aqp","fw_water","fw_sediment","ma_aqp","ma_sediment","ma_water"]

wtp = {}
for name in name_list:
    dict1 = {}
    for i in os.listdir(name):
        path = name +"/"+i

        if os.path.getsize(path) == 0:
            j = pd.DataFrame(columns=["ARGs"])
            dict1[i.strip(".sam.abn")] = j
        if os.path.getsize(path) > 0:
            j = pd.read_csv(path,sep="\t", names=["ARGs",i.strip(".sam.abn")])
            dict1[i.strip(".sam.abn")] = j
    temp1 = pd.DataFrame(columns=['ARGs'])
    for i in dict1:

        temp1 =  pd.merge(temp1, dict1[i], left_on=['ARGs'], right_on=['ARGs'], how='outer')
        temp1 = temp1.fillna(0)
        
        temp1['Length'] = temp1['ARGs'].str.extract(r'\|(\d+-\d+)\|')

        temp1['Length'] = temp1['Length'].str.split('-').apply(lambda x: int(x[1]) - int(x[0]) if isinstance(x, list) and not pd.isna(x).all() else 0)
        temp1.insert(loc=1, column='Length', value=temp1.pop('Length'))
    wtp[name] = temp1
    
```

```python
#merging and creating 3 sub categories to make marine and freshwater dataframes 

#formerging
x = wtp["fw_aqp"]
y = wtp["fw_sediment"]
z = wtp["fw_water"]

merged_df =  pd.merge(x, y, left_on=['ARGs'], right_on=['ARGs'], how='outer')
# merged_df
merged_df2 =  pd.merge(merged_df, z, left_on=['ARGs'], right_on=['ARGs'], how='outer')
# merged_df2
merged3 = merged_df2.fillna(0)

freshwater1 = merged3


#formerging
x = wtp["ma_aqp"]
y = wtp["ma_sediment"]
z = wtp["ma_water"]

merged_df =  pd.merge(x, y, left_on=['ARGs'], right_on=['ARGs'], how='outer')
# merged_df
merged_df2 =  pd.merge(merged_df, z, left_on=['ARGs'], right_on=['ARGs'], how='outer')
# merged_df2
merged3 = merged_df2.fillna(0)

marine1 = merged3
```

```python
#adjusting the length column in each dataframes 

freshwater1['Length'] = freshwater1['ARGs'].str.extract(r'\|(\d+-\d+)\|')

freshwater1['Length'] = freshwater1['Length'].str.split('-').apply(lambda x: int(x[1]) - int(x[0]) if isinstance(x, list) and not pd.isna(x).all() else 0)
freshwater1.insert(loc=1, column='Length', value=freshwater1.pop('Length'))
freshwater1.pop('Length_x')
freshwater1.pop('Length_y')

# freshwater1.to_csv("freshwater1.csv")

# freshwater1.head()



marine1['Length'] = marine1['ARGs'].str.extract(r'\|(\d+-\d+)\|')

marine1['Length'] = marine1['Length'].str.split('-').apply(lambda x: int(x[1]) - int(x[0]) if isinstance(x, list) and not pd.isna(x).all() else 0)
marine1.insert(loc=1, column='Length', value=marine1.pop('Length'))
marine1.pop('Length_x')
marine1.pop('Length_y')
# marine1.to_csv("marine1.csv")
marine1
```
